```python
##### SEU CODIGO COMECA AQUI #####
import pandas as pd
books = pd.read_csv("books.csv", sep = ',',on_bad_lines='skip')
print(books.shape)
print(books.describe())
```

```python
##### SEU CODIGO COMECA AQUI #####
print(books.isna().sum())
print(books.duplicated().sum())
```

```python
##### SEU CODIGO COMECA AQUI #####
books = books.drop(["isbn","isbn13","text_reviews_count","publication_date","publisher","bookID"], axis = 1)
books.head()
```

# Colunas

* title = títulos dos livros
* authors = autores dos livros
* average_rating = nota média dos livros
* language_code = língua que o livro está escrito
* num_pages = quantidade de páginas que o livro tem
* ratings_count = quantidade de pessoas que deram nota ao livro

```python
##### SEU CODIGO COMECA AQUI #####
filtro_nota = books[(books["average_rating"] > 4.55) & (books["ratings_count"] > 100000)]
print(filtro_nota)
```

```python
##### SEU CODIGO COMECA AQUI #####
filtro_notas_ruim = books[(books["average_rating"] < 3) &(books["ratings_count"] > 5000)]
filtro_notas_ruim
```

```python
##### SEU CODIGO COMECA AQUI #####
media_nota_boa= filtro_nota['  num_pages'].mean()
media_nota_ruim = filtro_notas_ruim['  num_pages'].mean()
if(media_nota_boa > media_nota_ruim):
    print(f'A média de página dos três livros mais bem avaliados ({media_nota_boa:.2f}) é maior que a média de páginas dos três livros mais mal avaliados ({media_nota_ruim:.2f})')
elif(media_nota_boa < media_nota_ruim):
    print(f'A média de página dos três livros mais bem avaliados ({media_nota_boa:.2f}) é menor que a média de páginas dos três livros mais mal avaliados ({media_nota_ruim:.2f})')
else:
    print(f'A média de página dos três livros mais bem avaliados ({media_nota_boa:.2f}) é igual que a média de páginas dos três livros mais mal avaliados ({media_nota_ruim:.2f})')
```
