## Condicionais

Condicionais sÃ£o estruturas de programaÃ§Ã£o que executam aÃ§Ãµes a partir de correponsdÃªncia de certas condiÃ§Ãµes.

## If
If ("Se" em portuguÃªs) Ã© uma estrura que executa uma aÃ§Ã£o caso certas condiÃ§Ãµes determinadas sejam atendidas.

## Else
Else ("Se nÃ£o" em portuguÃªs) Ã© um estrutura que executa uma aÃ§Ã£o caso certas condiÃ§Ãµes determinadas estabelecidas anteriormente (if) <b>nÃ£o</b> sejam atendidas. 

## Elif
Elif (Else + If) Ã© uma estrura que executa uma aÃ§Ã£o caso certas condiÃ§Ãµes <b>distintas</b> daquelas determinadas anteriormente sejam atendias.


## And
And ("e" em portuguÃªs) Ã© uma estrutura que permite determinar duas ou mais condiÃ§Ãµes que devem ser simultÃ¢neamente atendidas para a execuÃ§Ã£o de uma aÃ§Ã£o (Se x < 0 <b>e</b> y = 0:).

## Or
Or ("ou" em portuguÃªs) Ã© uma estrutura que permite determinar duas ou mais condiÃ§Ãµes de tal modo que no mÃ­nimo uma delas deve ser atendida para a execuÃ§Ã£o de uma aÃ§Ã£o (Se x < 0 <b>ou</b> y = 0:)

## Exemplo:
Um fazendeiro possui uma fazenda de tomates e deseja vende-los de acordo com seu tamanho e maturaÃ§Ã£o. Se os tomates forem grandes(g) e maduros ele os vendera por 2,00 reais a unidade. Se os tomates forem pequenos e maduros ou grandes e verdes ele deve os vender por 1,00 real a unidade. Caso o tomate for pequeno e verde ele deve os descartar.

```python
print("Qualidade do Tomate")
MaturaÃ§Ã£o = input("MaturaÃ§Ã£o(v/m): ")
Tamanho = input('Tamanho(p/g): ')

if MaturaÃ§Ã£o == "m" and Tamanho == "g":
    print("Vender tomate a R$2,00.")
elif MaturaÃ§Ã£o == "m" and Tamanho == "p" or MaturaÃ§Ã£o == "v" and Tamanho == "g":
    print("Vender tomate a R$1,00")
else:
    print("Descartar tomate.")
```
